```python
def get_score(url):
    if url == "http://money.cnn.com/2017/04/04/news/economy/obamacare-tennessee/index.html":
        return 2
    elif url == "http://www.breitbart.com/big-government/2017/04/04/trump-town-hall-common-core-end-it/":
        return 5
    else:
        return 10 
```

Given a url, determine if the url is fake news / biased or not. A score will be given in the range of 1-10. A score of 1 signifies that the article is reputable. A score of 10 signifies that the article contains false statements. 

```python
# example of a source that is reputable 

# title: Places where Obamacare could die
print get_score("http://money.cnn.com/2017/04/04/news/economy/obamacare-tennessee/index.html")
```

```python
# example of a source that is semi - reputable 

# ‘Common Core… We Have to End It’
print get_score("http://www.breitbart.com/big-government/2017/04/04/trump-town-hall-common-core-end-it/")
```

```python
# example of a source that is fake news

# BUSTED! CLINTON ADVISOR JOHN PODESTA Paid BIG Money By Russia
print get_score("http://100percentfedup.com/busted-clinton-advisor-john-podesta-paid-big-money-by-russia-video/")
```
