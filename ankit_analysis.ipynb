```python
import nltk

#fakenews=fake[fake["type"]=="fake"]
fake_e=fake[fake["language"]=="english"]

from sklearn.model_selection import train_test_split
train,test = train_test_split(fake_e,test_size=0.15)
```

```python
from nltk.corpus import stopwords 

train.columns.values
```

```python
train.head()
```

```python
import warnings
warnings.filterwarnings("ignore")


import re
def refined(i):
    letters = re.sub("[^a-zA-Z]", " ", i) 
    words = letters.lower().split()
    stops = set(stopwords.words("english"))
    useful = [w for w in words if not w in stops]
    return( " ".join( useful ))


train["text"].fillna(" ",inplace=True)    
train["text"] = train["text"].apply(refined)
train["author"].fillna(" ",inplace=True)    
train["author"] = train["author"].apply(refined)
train["site_url"].fillna(" ",inplace=True)    
train["site_url"] = train["site_url"].apply(refined)
train["title"].fillna(" ",inplace=True)    
train["title"] = train["title"].apply(refined)
train["thread_title"].fillna(" ",inplace=True)    
train["thread_title"] = train["thread_title"].apply(refined)
train2 = train.copy()
train.head()
```

```python
train_t = train2.copy()
from sklearn.feature_extraction.text import CountVectorizer
vectorize = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 5000) 
train_t["text"] = vectorize.fit_transform(train_t["text"]).toarray()
train_t["author"] = vectorize.fit_transform(train_t["author"]).toarray()
train_t["site_url"] = vectorize.fit_transform(train_t["site_url"]).toarray()
train_t["title"] = vectorize.fit_transform(train_t["title"]).toarray()
train_t["thread_title"] = vectorize.fit_transform(train_t["thread_title"]).toarray()
train_t.describe()
```

```python
print(sum(pd.isnull(train_t["domain_rank"])))
print(sum(pd.isnull(test["domain_rank"])))
#as we see we have 3546 null values
```

```python
#We can fill these null values by taking the median 

train_t["domain_rank"]=train_t["domain_rank"].fillna(train_t.domain_rank.median(axis=0))
test["domain_rank"]=test["domain_rank"].fillna(test.domain_rank.median(axis=0))
```

```python
# Adding a column that puts in the spam score ranking
train_t["IsSpam"]=np.sign(train_t["spam_score"]-0.5)
train_t.head()
```

```python
train_t.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree=DecisionTreeClassifier(min_samples_split=2,random_state=7)
dtree_feat=train_t[["text", "author", "site_url", "title", "thread_title","domain_rank"]].values
dtree_fit=dtree.fit(dtree_feat,train_t["IsSpam"])
```

```python
target=train_t["IsSpam"].values
```

```python
dtree_fit.score(dtree_feat,target)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier(min_samples_split=2, n_estimators=100, random_state=7)
```

```python
rfc_feat=train_t[["text", "author", "site_url", "title", "thread_title","domain_rank"]].values
```

```python
rfc_fit=rfc.fit(rfc_feat,train_t["IsSpam"])
```
